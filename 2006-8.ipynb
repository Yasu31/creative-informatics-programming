```python
import numpy as np
import string
from anytree import Node, RenderTree, PreOrderIter, PostOrderIter, Walker
from anytree.search import find
from anytree.exporter import DotExporter
```

```python
def caesarShift(text, key):
    returnText = ""
    text = text.lower()
    for letter in text:
        if letter in string.ascii_letters:
            returnText+=chr((ord(letter)+key-ord('a'))%26+ord('a'))
        else:
            returnText += letter
    return returnText
```

```python
print(caesarShift("xyzabc!.", -1))
print(caesarShift("ibm", 25))
print(caesarShift("IBM", 25))
```

```python
def characterCount(text):
    # store how many times each letter appeared
    counter = [0]*26
    for letter in text:
        if 0<= ord(letter)-ord('a')<26:
            counter[ord(letter)-ord('a')] +=1
        if 0<= ord(letter)-ord('A')<26:
            counter[ord(letter)-ord('A')] +=1
    sortedArgs = np.argsort(counter)[::-1]
    i = 0
    print("letter\tappearance(times)")
    for arg in sortedArgs:
        print("{}\t{}".format(chr(arg+ord('a')), counter[arg]))
    return counter
```

```python
def encode(text):
    '''
    encode the text with a random mapping, and save to file.
    '''
    mapChar = np.arange(26)
    np.random.seed(0)
    np.random.shuffle(mapChar)
    print("mapping is: {}".format(mapChar))
    returnText = ""
    for letter in text:
        if letter in string.ascii_lowercase:
            returnText += chr(mapChar[ord(letter)-ord('a')]+ord('a'))
        elif letter in string.ascii_uppercase:
            returnText += chr(mapChar[ord(letter)-ord('A')]+ord('A'))
        else:
            returnText += letter
    return returnText
```

```python
with open('resources/2006-8/plaegeuius.txt') as f:
    text = f.read().replace("\n",'')
print(encode(text))
```

```python
# based on result from before, create guesses for mapping

def decodeWithMapping(text, charMap):
    '''
    charMap is a dictionary with key -> value pairs of code -> decoded
    '''
    text = text.lower()
    decodedText = ''
    for letter in text:
        if not letter in string.ascii_letters:
            decodedText += letter
            continue
        try:
            decodedText += charMap[letter]
        except KeyError:
            decodedText += '_'
    return decodedText
```

```python
with open('resources/2006-8/beemovie.txt') as f:
    text = f.read().replace("\n",' ')
encodedText = encode(text)
count = characterCount(encodedText)
```

```python
def id2letter(arg):
    if arg == 95:
        letter = "NL"
    elif arg == ord(' ')-32:
        letter = "SP"
    else:
        letter = chr(arg+32)
    return letter
def characterCountAll(text):
    # store how many times each letter appeared, not just letters
    counter = [0]*96
    for letter in text:
        asciicode = ord(letter)
        if 32 <= asciicode <= 126:
            counter[asciicode-32] += 1
        elif asciicode == ord('\n'):
            counter[95] += 1
    sortedArgs = np.argsort(counter)[::-1]
    i = 0
    print("letter\tappearance(times)")
    for arg in sortedArgs:
        letter = id2letter(arg)
        print("{}\t{}".format(letter, counter[arg]))
    return counter
```

```python
with open('resources/2006-8/beemovie.txt','r') as f:
    text = f.read()
count = characterCountAll(text)
orphanNodes = []
for i in range(len(count)):
    orphanNodes.append(Node(id2letter(i), count=count[i]))
while len(orphanNodes) > 1:
    smallestIndex = np.argmin([node.count for node in orphanNodes])
    smallestNode = orphanNodes.pop(smallestIndex)
    smallestNode.lr = '1'
    secondSmallestIndex = np.argmin([node.count for node in orphanNodes])
    secondSmallestNode = orphanNodes.pop(secondSmallestIndex)
    secondSmallestNode.lr = '0'
#     print("orphan node with smallest value is: {}".format(smallestNode))
#     print("orphan node with second smallest value is: {}".format(secondSmallestNode))
    newNode = Node('', count = smallestNode.count+secondSmallestNode.count, children=[secondSmallestNode, smallestNode])
    orphanNodes.append(newNode)
top = orphanNodes[0]
print(RenderTree(top))
w = Walker()
charEncodingDict = {}
for i in range(len(count)):
    letter = id2letter(i)
    node = find(top, lambda node: letter == node.name)
    path = w.walk(top, node)[2]
    value = ''
    for step in path:
        value += step.lr
    charEncodingDict[letter] = value
    print("{}\t{}".format(letter, value))
```

```python
bits_sum = 0
for i in range(len(count)):
    letter = id2letter(i)
    bits_sum += count[i] * len(charEncodingDict[letter])
average = bits_sum / sum(count)
print("bits\t{}\nbytes\t{}\naverage bits per letter\t{}".format(bits_sum, bits_sum//8+1,average))
```
